```python
# Importing necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading Applications Raw Data
data = pd.read_csv("Code1_application.csv")
data.head()
```

```python
# Data After creating new variables
df=pd.read_csv('Code2_All_Variable_Data.csv')
df.head()
```

```python
# Adding  fraud labels
df['fraud_label'] = data['fraud_label']
df.head()
```

```python
# retaining only relevant 29 columns
col = ['record','fulladdress_daysince','fulladdress_vel_30','fulladdress_vel_14','fulladdress_vel_7','fulladdress_vel_3',
        'fulladdress_relVel_14','fulladdress_relVel_30','fulladdress_relVel_60','fulladdress_vel_1','fulladdress_phone_daysince',
        'fulladdress_phone_vel_30','namedob_daysince','namedob_vel_30','ssn_daysince','ssn_vel_30',
        'ssn_namedob_daysince','ssn_namedob_vel_30','fulladdress_phone_vel_60','ssn_name_daysince','ssn_name_vel_30','namedob_vel_60',
        'ssn_namedob_vel_60','fulladdress_phone_vel_14','ssn_name_vel_60','namedob_vel_14','ssn_dob_vel_14','ssn_vel_14','fraud_label']

df = df[col]
df.head()
col
```

```python
# Updating days since
df2 = pd.read_csv("Code2_DaysSince.csv",index_col = 0)
df2.head()
```

```python
# Changing names to previous style
name_mapping = {
                'record': 'record',
                'fraud_label': 'fraud_label',
    'fulladdress_vel_1': 'vel_fulladdress_1',
 'fulladdress_vel_3': 'vel_fulladdress_3',
 'fulladdress_vel_7':'vel_fulladdress_7',
 'fulladdress_vel_14': 'vel_fulladdress_14',
 'fulladdress_vel_30': 'vel_fulladdress_30',
 'fulladdress_relVel_14': 'relv_fulladdress_14',
 'fulladdress_relVel_30': 'relv_fulladdress_30',
 'fulladdress_relVel_60': 'relv_fulladdress_60',
 'fulladdress_phone_vel_14': 'vel_fulladdressphone_14',
 'fulladdress_phone_vel_30': 'vel_fulladdressphone_30',
 'fulladdress_phone_vel_60': 'vel_fulladdressphone_60',
 'ssn_vel_14': 'vel_ssn_14', 
 'ssn_vel_30': 'vel_ssn_30',
 'ssn_namedob_vel_30': 'vel_ssnnamedob_30',
 'ssn_namedob_vel_60': 'vel_ssnnamedob_60',
 'ssn_name_vel_30': 'vel_namessn_30',
 'ssn_name_vel_60': 'vel_namessn_60',
 'ssn_dob_vel_14': 'vel_ssndob_14',
 'namedob_vel_14': 'vel_namedob_14',
 'namedob_vel_30': 'vel_namedob_30',
 'namedob_vel_60': 'vel_namedob_60',
 'namedob_daysince' : 'dayssince_namedob',
 'ssn_name_daysince' :  'dayssince_namessn',
 'ssn_namedob_daysince': 'dayssince_ssnnamedob',
 'ssn_daysince': 'dayssince_ssn',
 'fulladdress_phone_daysince': 'dayssince_fulladdressphone',
 'fulladdress_daysince': 'dayssince_fulladdress'
 
}
pd.set_option('display.max_columns', None) 
df.columns = df.columns.map(name_mapping)
df.head()
```

```python
# Updating dayssince in right format
df['dayssince_namedob'] = df2['dayssince_namedob']
df[ 'dayssince_namessn']= df2[ 'dayssince_namessn']
df['dayssince_ssnnamedob'] = df2['dayssince_ssnnamedob']
df['dayssince_ssn'] = df2['dayssince_ssn']
df['dayssince_fulladdressphone'] = df2['dayssince_fulladdressphone']
df['dayssince_fulladdress'] = df2['dayssince_fulladdress']
df.head()
```

```python
list(df.columns)
```

```python

days_since_columns = ['dayssince_namedob','dayssince_namessn','dayssince_ssnnamedob',
                      'dayssince_ssn','dayssince_fulladdressphone','dayssince_fulladdress']

for i in days_since_columns: 
    maximum = df[i].max()
    df[i] = maximum - df[i]
df.head()
```
