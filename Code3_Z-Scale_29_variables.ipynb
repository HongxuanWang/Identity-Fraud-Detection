```python
# Importing necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
data=pd.read_csv('Code3_Final_29_Variable_Data.csv',index_col = 0)
data.head()
```

```python
# Z-scale
df2 = (data.iloc[:,1:28] - data.iloc[:,1:28].mean()) / data.iloc[:,1:28].std()
df2.head()
```

```python
# Adding record and fraud label
df2['fraud_label'] = data['fraud_label']
df2['record'] = data['record']
cols = list(df2.columns)
cols = [cols[-1]] + cols[:-1]
df2 = df2[cols]
df2.head()
```
